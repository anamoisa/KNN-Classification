```
#X= df_pokemon_non_legends[['Type 1','HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','Generation']]

modelNN = NearestNeighbors(n_neighbors=1).fit(df_pokemon_non_legends.iloc[:,1:9])
modelNN
```

```
legendary_pokemons = list(df_pokemon_legends['Name'])
replacements = []

for i in legendary_pokemons:
  X = np.array(df_pokemon_legends[df_pokemon_legends['Name']== i])[:,1:9] 
  how_to_search = modelNN.kneighbors(X)
  closest_neighbor = int(how_to_search[1])
  closest_pokemon = df_pokemon_non_legends[df_pokemon_non_legends.index == closest_neighbor].iloc[0,0] 
  print('The non-legendary pokemon substitude for',i, "is", closest_pokemon)
  replacements.append(closest_pokemon) 

```
